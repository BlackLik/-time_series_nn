```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
PATH = "VTBR_200201_201219.csv"
df = pd.read_csv(PATH)
df['<DATE>'] = pd.to_datetime(df['<DATE>'])
df['date_delta'] = (df['<DATE>'] - df['<DATE>'].min()) / np.timedelta64(1, "D")
```

```python
df
```

```python
df.info()
```

```python
def draw(X, y):
    X = np.array(X)
    plt.figure(figsize=(10, 5))
    plt.plot(y, X)
    plt.ylabel("Price")
    plt.xlabel('Date')
    plt.show()
```

```python
draw(df['<OPEN>'], df['date_delta'])
```

```python
window_of_size = window_size = 20
```

```python
train = df.loc[:window_size+window_of_size, :]
test = df.loc[window_size+window_of_size+1:, :]
```

```python
train.head()
```

```python
test.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(train['<DATE>'], train['<OPEN>'], label="Train")
plt.plot(test['<DATE>'], test['<OPEN>'], label="Test")
plt.ylabel("Price")
plt.xlabel('Date')
plt.legend(loc="best")
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import machin
model = machin.NetRegressor(n_input=window_size)
```

```python
y_hat = train.copy()
```

```python
for i in range(window_of_size):
    if i == 0:
        n_input = train[i:i+window_size]['<OPEN>'].values
    else:
        n_input = np.vstack((n_input, train[i:i+window_size]['<OPEN>'].values))
```

```python
n_output = train['<OPEN>'][window_size:window_size+window_of_size].values.reshape(-1, 1)
```

```python
n_output.T
```

```python
model.fit(n_input, n_output)
```

```python
new_base = train[0:0+window_size]['<OPEN>'].values
```

```python
model.predict(new_base)[0]
```

```python
train['<OPEN>'][0+i+window_size]
```

```python
z_hat = train.copy()
```

```python
reports = []
for i in range(0, len(train) - window_size - window_of_size):
    for j in range(i, i+window_of_size):
        if j == i:
            n_input = train[j:j+window_size]['<OPEN>'].values
        else:
            n_input = np.vstack((n_input, train[j:j+window_size]['<OPEN>'].values))
    n_output = train['<OPEN>'][i+window_of_size:i+window_of_size+window_of_size].values.reshape(-1, 1)
    model.fit(n_input, n_output)
    
    new_base = train[i+window_of_size:i+window_of_size+window_size]['<OPEN>'].values
    y_hat.loc[i+window_of_size+window_size, '<OPEN>'] = model.predict(new_base)[0]
    report = abs((y_hat['<OPEN>'][i+window_of_size+window_size] -  train['<OPEN>'][i+window_of_size+window_size])/train['<OPEN>'][i+window_of_size+window_size])*100
    reports.append(report)
    
    if report < 30:
        z_hat.loc[i+window_of_size+window_size, '<OPEN>'] = np.NaN
    else:
        print(report)
    
reports = np.array(reports)
```

```python
for i in range(0, window_size+window_of_size):
    z_hat.loc[i, '<OPEN>'] = np.NaN
```

```python
plt.figure(figsize=(10, 5))
plt.plot(train['<DATE>'], train['<OPEN>'], label="Train")
plt.plot(y_hat['<DATE>'], y_hat['<OPEN>'], label="NN")
plt.plot(z_hat['<DATE>'], z_hat['<OPEN>'], 'rx', label="Anomalies")
plt.ylabel("Price")
plt.xlabel('Date')
plt.legend(loc="best")
plt.title("NN")
plt.show()
```

```python
test = test[:].reset_index(drop=True)
z_hat_test = test.copy()
y_hat_test = test.copy()
```

```python
reports = []
for i in range(0, len(test) - window_of_size - 1):
    new_base = test[i:i+window_size]['<OPEN>'].values
    y_hat_test.at[i+window_size, 1] = model.predict(new_base)[0]
    report = abs((np.array(y_hat_test['<OPEN>'][i+window_size]) -  np.array(test['<OPEN>'][i+window_size]))/np.array(test['<OPEN>'][i+window_size]))*100
    reports.append(report)
    if report < 30:
        z_hat_test.at[i+1+window_size, '<OPEN>'] = np.NaN
    else:
        print(report)

reports = np.array(reports)
for i in range(0, window_size+1):
    z_hat_test.loc[i, '<OPEN>'] = np.NaN
```

```python
df_predict
```

```python
z_hat_predict = df_predict.copy()
y_hat_predict = df_predict.copy()
```

```python
reports = []
for i in range(0, len(df_predict) - window_size ):
    new_base = df_predict[i:i+window_size]['<OPEN>'].values
    y_hat_predict.at[i+window_size, 1] = model.predict(new_base)[0]
    report = (abs((np.float32(y_hat_predict['<OPEN>'][i+window_size]) -  np.float32(df_predict['<OPEN>'][i+window_size]))/np.float32(df_predict['<OPEN>'][i+window_size]))*100)
    reports.append(report)
    if report < 30:
        z_hat_predict.at[i+1+window_size, '<OPEN>'] = np.NaN
    else:
        print(report)

reports = np.array(reports)

for i in range(0, window_size+1):
    z_hat_predict.at[i, '<OPEN>'] = np.NaN
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_predict['<DATE>'], df_predict['<OPEN>'], label="Predict")
plt.plot(y_hat_predict['<DATE>'], y_hat_predict['<OPEN>'], label="NN")
plt.plot(z_hat_predict['<DATE>'], z_hat_predict['<OPEN>'], 'rx', label="Anomalies")
plt.ylabel("Price")
plt.xlabel('Date')
plt.legend(loc="best")
plt.title("NN")
plt.show()
```
